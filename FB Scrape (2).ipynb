```python
names= []
profile_links = []
image_links = []
images_info = []
post_description=[]
```

```python
while True:
    
        for element in driver.find_elements(by=By.XPATH,value="//div[@class='x1ja2u2z xh8yej3 x1n2onr6 x1yztbdb']"):

            try:
                name = driver.find_element(by=By.XPATH, value='.//h2//span/a/strong/span').text
                print('name: ',name)
                names.append(name)
            except:
                name= names[0]
                print('name: ',name)
                names.append(name)
            time.sleep(3)

            try:
                profile_link = driver.find_element(by=By.XPATH, value='.//h2/span/a[1]').get_attribute('href')
                print('prof link: ',profile_link)
                profile_links.append(profile_link)
            except:
                profile_link = profile_links[0]
                print('prof link: ',profile_link)
                profile_links.append(profile_link)
            time.sleep(3)

            try:
                image_link = element.find_element(by=By.XPATH, value='.//div[1]/div/a/div[1]/div/div/div/img').get_attribute('src')
                image_info = element.find_element(by=By.XPATH, value='.//div[1]/div/a/div[1]/div/div/div/img').get_attribute('alt')
                image_links.append(image_link)
                print('img link: ',image_link)
                images_info.append(image_info)
                print('img info: ',image_info)
            except:
                image_links.append('NaN')
                images_info.append('NaN')
                print('img link: ','NaN')
                print('img link: ','NaN')
            time.sleep(3)  

            try:
                description = element.find_element(by=By.XPATH, value='.//div[@data-ad-comet-preview="message"]/div/div//div/div').text
                post_description.append(description)
                print('desc: ',description)
            except:
                post_description.append('NaN')
                print('desc: ','NaN')  
            print('--------------------------------------------------------------------')

        time.sleep(5)
        
        
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)
```

```python
print("Names:", len(names))
print("Prof_Links:", len(profile_links))
print("Img_Links:", len(image_links))
print("Img_Info:", len(images_info))
print("Post_Desp:",len(post_description))
```
